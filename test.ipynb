```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create folder for saving graphs
os.makedirs('./outputs', exist_ok=True)

# Load data
print("Loading data for visualization...")
df_imdb = pd.read_csv('./data/imdb_processed.csv').dropna(subset=['clean_text'])
df_twit = pd.read_csv('./data/twitter_processed.csv').dropna(subset=['clean_text'])

# Add tags for graphing
df_imdb['Source'] = 'IMDB'
df_twit['Source'] = 'Twitter'
df_all = pd.concat([df_imdb, df_twit], ignore_index=True)

# Map numeric sentiment back to words for pretty graphs
df_all['Sentiment_Label'] = df_all['sentiment'].map({1: 'Positive', 0: 'Negative'})

# Set visual style
sns.set_theme(style="whitegrid")

# ---------------------------------------------------------
# Graph 1: Class Distribution (Are our datasets balanced?)
# ---------------------------------------------------------
plt.figure(figsize=(10, 6))
sns.countplot(data=df_all, x='Source', hue='Sentiment_Label', palette='viridis')
plt.title("Dataset Balance: Positive vs Negative Sentiment", fontsize=16)
plt.ylabel("Number of Samples")
plt.savefig('./outputs/sentiment_distribution.png', bbox_inches='tight')
plt.show()
```

```python
# ---------------------------------------------------------
# Graph 2: Feature Engineering - Word Count Distributions
# ---------------------------------------------------------
# We limit to 500 words to cut off massive outlier reviews
plt.figure(figsize=(12, 6))
sns.histplot(data=df_all[df_all['word_count'] < 500], x='word_count', hue='Source', 
             kde=True, bins=50, palette='magma', alpha=0.6)
plt.title("Feature Engineering: Word Count Distribution (IMDB vs Twitter)", fontsize=16)
plt.xlabel("Word Count")
plt.ylabel("Frequency")
plt.savefig('./outputs/word_count_distribution.png', bbox_inches='tight')
plt.show()
```

```python
# ---------------------------------------------------------
# Graph 3: Feature Engineering - Punctuation Intensity
# ---------------------------------------------------------
plt.figure(figsize=(10, 6))
sns.barplot(data=df_all, x='Sentiment_Label', y='exclamation_count', hue='Source', palette='Set2')
plt.title("Feature Engineering: Average Exclamation Marks per Sentiment", fontsize=16)
plt.ylabel("Avg Exclamation Marks (!)")
plt.savefig('./outputs/exclamation_impact.png', bbox_inches='tight')
plt.show()

print("Graphs successfully generated and saved to the 'outputs' folder!")
```
